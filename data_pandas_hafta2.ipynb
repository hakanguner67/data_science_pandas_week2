1 - Tum ogrencilerin bilgilerini tek bir dataframe icinde su sutunlardan olusacak sekilde birlestiriniz; > " Sinif ismi, Ogrenci Adi, Dogru Sayisi, Yanlis Sayisi, Bos Sayisi" 

```python
import pandas as pd
import numpy as np
#ayni dizinde oldugumuz icin ekstra yol girilmedi
sayfa = pd.ExcelFile("py_mind.xlsx")
sayfa1 = pd.ExcelFile("py_opinion.xlsx")
sayfa2 = pd.ExcelFile("py_sense.xlsx")
sayfa3 = pd.ExcelFile("py_science.xlsx")

files = [sayfa, sayfa1, sayfa2 , sayfa3]
siniflar=[]
isimler=[]
dogru_sayisi=[]
yanlis_sayisi=[]
bos_sayisi=[]

for i in files:                                     
    for j in range(len(i.sheet_names)):                
        
        data = pd.read_excel(i, i.sheet_names[j]) 
        
        isimler += [data.columns[0]]                
        dogru_sayisi += [data.iloc[-3,1]]
        yanlis_sayisi += [data.iloc[-2,1]]
        bos_sayisi += [data.iloc[-1,1]]
        
        
        if i == sayfa:
            sinif = "py_mind"
        elif i == sayfa1:
            sinif = "py_opinion"
        elif i == sayfa2:
            sinif = "py_sense"
                
        elif i == sayfa3:
            sinif = "py_science"
       
        siniflar += [sinif]
        

    
Ogrenciler = pd.DataFrame({"Sinif": siniflar, "Ogrenci Adi": isimler ,"Dogru": dogru_sayisi,"Yanlis": yanlis_sayisi, "Bos": bos_sayisi})

        
Ogrenciler.index = Ogrenciler.index + 1  # sifirdan baslamamasi icin 
for a in Ogrenciler: # 24 . siradaki selcuk d nin dogru sayilari ve yanlis sayilari girilmediginden onlari sifir olarak gosterdik.
    for i in Ogrenciler.Dogru :
        if i == "B":
            Ogrenciler.Dogru.replace("B",0, inplace=True) 
    for j in Ogrenciler.Yanlis:
        if j == "A":
            Ogrenciler.Yanlis.replace("A",0, inplace=True)
    for k in Ogrenciler.Bos:
        if k == "A":
            Ogrenciler.Bos.replace("A",0, inplace=True)
Ogrenciler
```
